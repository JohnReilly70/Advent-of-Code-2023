```python
result = 0
temp_num_list = []
temp_first_last_num = 0
```

```python
with open("Day1_Input.txt") as file:
    for line in file: #Looping through each line in the file
        for char in line: #Looping through each char in the line
            if char.isnumeric(): #checking if char is a number
                temp_num_list.append(char) #append to list if is a number
        temp_first_last_num = int(temp_num_list[0] + temp_num_list[-1]) #Taking the first and last value of the list which are strings, adding them adds them like a string. Using Int() then converts to number which allows the math operation to wokr
        result = result + temp_first_last_num #Adding the value to the result
        temp_num_list = [] #clearing the temp list for the next line
    temp_first_last_num = 0 #clearing temp first last num as it is a new line.
```
