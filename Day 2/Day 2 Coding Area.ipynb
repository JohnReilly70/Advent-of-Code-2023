```python
#General Variables
Game_Number_Int = 0
Round_Number_Int = 0
Round_Success_Count = 0
Game_Match_Counter_Int = 0
Fail_Count_Int = 0
Game_List = []
Round_List = []
Counter_Dict = {"red":0, "green":0, "blue":0}
Largest_Colour_Value_Dict = {"red":0, "green":0, "blue":0}
match = 0
power = 1
total_power = 0

#Open File
#Read each Line from File
with open("Day2_Input_Part1.txt") as file:
    for line in file: # Looping through each line
        line = line.strip()
        Game_List= line.split(":")
        Game_Number_Int = Game_List[0][4:]
        print(f"Game Num:{Game_Number_Int}")
        # print(Game_List)
        Round_List = Game_List[1].split(";")
        # print(Round_List)
        for round in Round_List:
            round = round.split(",")
            # print(round)
            for colours in round:
                colours = colours.lstrip().split(" ")
                Counter_Dict[colours[1]] += int(colours[0])
                print(Counter_Dict)
                if Counter_Dict["red"] > 12 or Counter_Dict["green"] > 13 or Counter_Dict["blue"] > 14:
                    Fail_Count_Int += 1
                    print(Counter_Dict)

            for key,val in Counter_Dict.items():
                print(key, val)
                if Largest_Colour_Value_Dict[key] < Counter_Dict[key]:
                    Largest_Colour_Value_Dict[key] = val
            print(f"End Round{Largest_Colour_Value_Dict}")
            Counter_Dict = {"red":0, "green":0, "blue":0}
        if Fail_Count_Int == 0:
            print("Enter Fail Count Int")
            Game_Match_Counter_Int += int(Game_Number_Int)
            match += 1
        for val in Largest_Colour_Value_Dict.values():
            power = power * val
            print(power)
        total_power += power
        power = 1
        Largest_Colour_Value_Dict = {"red":0, "green":0, "blue":0}
        Fail_Count_Int = 0
```
