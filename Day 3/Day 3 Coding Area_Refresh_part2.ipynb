```python
symbols = ["*", "$", "#","+","@","-","-","/","=","%","&"]
```

```python


part_num_string = ""
counted_part_num_list = []
line_update_list = []
counter = 0
updated_file_list = []
line_coord_only_list = []
start_coord = []
symbol_coord_dict = {}
complete_coord_only_list = []
symbols_coord = []


#Open Example file
with open("Day3_Input_Part2.txt") as file:



    #loop through each line
    for index_line, line in enumerate(file):
        line_update_list = []
        line = line.strip()
        for index_char, char in enumerate(line.rstrip()):
            print(f"index:{index_char}:len(line){len(line)}")
            if char.isnumeric():
                if not part_num_string:
                    start_coord = [index_line,index_char]
                part_num_string += char
                if index_char == len(line)-1:
                    for part_num_index in range(0,len(part_num_string)):
                        line_update_list.append([start_coord,part_num_string])
                    part_num_string = ""


            elif (not char.isnumeric()) or (index_char == len(line)):
                if not part_num_string:
                    line_update_list.append(char)
                if part_num_string:
                    for part_num_index in range(0,len(part_num_string)):
                        line_update_list.append([start_coord,part_num_string])
                    part_num_string = ""
                    line_update_list.append(char)
            if char in symbols:
                if char == "*":
                    symbols_coord.append([index_line,index_char])
        updated_file_list.append(line_update_list)
```
