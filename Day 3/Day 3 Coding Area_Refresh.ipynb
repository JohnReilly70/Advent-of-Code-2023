```python
symbols = ["*", "$", "#","+","@","-","-","/","=","%","&"]
```

```python


part_num_string = ""
counted_part_num_list = []
line_update_list = []
counter = 0
updated_file_list = []
line_coord_only_list = []
start_coord = []
symbol_coord_dict = {}
complete_coord_only_list = []
symbols_coord = []


#Open Example file
with open("Day3_Input_Part1.txt") as file:



    #loop through each line
    for index_line, line in enumerate(file):
        line_update_list = []
        line = line.strip()
        for index_char, char in enumerate(line.rstrip()):
            if char.isnumeric():
                if not part_num_string:
                    start_coord = [index_line,index_char]
                part_num_string += char
                if index_char == len(line)-1:
                    for part_num_index in range(0,len(part_num_string)):
                        line_update_list.append([start_coord,part_num_string])
                    part_num_string = ""


            elif (not char.isnumeric()) or (index_char == len(line)):
                if not part_num_string:
                    line_update_list.append(char)
                if part_num_string:
                    for part_num_index in range(0,len(part_num_string)):
                        line_update_list.append([start_coord,part_num_string])
                    part_num_string = ""
                    line_update_list.append(char)
            if char in symbols:
                symbols_coord.append([index_line,index_char])
        updated_file_list.append(line_update_list)
```

```python
max_row_len = len(updated_file_list)
max_column_len = len(updated_file_list[0])
```

```python
for symbol in symbols_coord:
    for row in range(-1,2):
        for column in range(-1,2):
            if (symbol[0]+row >= 0 and symbol[0]+row < max_row_len):
                if (symbol[1]+column >= 0 and symbol[1]+column < max_column_len):
                        if isinstance(updated_file_list[int(symbol[0]+row)][int(symbol[1]+column)], list):
                            temp_list = updated_file_list[int(symbol[0]+row)][int(symbol[1]+column)]
                            counted_part_num_list[int(symbol[0]+row)].append(temp_list[0][1])
                else:
                    print("Failed on Column")
            else:
                print("Failed on Row")
```
